```python
import numpy as np
from scipy.stats import norm
```

```python
sample = [
    61, 58, 62, 59, 65, 60, 56, 63, 61, 64,
    55, 62, 59, 60, 63, 58, 61, 62, 60, 57,
    59, 64, 61, 63, 58, 60, 62, 59, 61, 66,
    57, 62, 60, 58, 63
]

population_mean = 60
population_std = 3
sample_mean = np.mean(sample)
n = len(sample)
alpha = 0.05     #----------------------> Significant value
```

```python
# calculating z-score
z_score = (sample_mean - population_mean)/(population_std/ np.sqrt(n))
z_score
```

```python
# Calculating p_value
p_value = 2 * (1 - norm.cdf(abs(z_score)))
p_value
```

```python
if p_value < alpha:
    print("Reject the null hypothesis")
else: 
    print("Accept the null hypothesis")
```
