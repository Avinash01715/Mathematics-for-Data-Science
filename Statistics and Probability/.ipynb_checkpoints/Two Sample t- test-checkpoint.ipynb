```python
import numpy as np
from scipy import stats
```

```python
# two samples
group_a = [78, 82, 85, 79, 81, 84, 80, 83, 77, 86]
group_b = [72, 75, 71, 74, 73, 76, 70, 78, 69, 74]

## null hypothesis ----> No significant difference in population mean of two groups
## alternate hypothesis ----> There is a significant difference in population mean of two groups

```

```python
t_stats, p_values = stats.ttest_ind(group_a, group_b, equal_var= False)  #------> This fn returns two values
print("t_stats= ", t_stats)
print("p_values= ", p_values)
```

```python
alpha = 0.05 #-------> significant value (given)

if p_values < alpha:
    print("We reject the null hypothesis")
else:
    print("We accept the null hypothesis")
```
