```python
#question-> is there a relationship between survival and gender in titanic.

## null hypothesis (Ho) ---> There is no relationship
## alternate hypothesis (H1) ---> There is a relationship
```

```python
import numpy as np 
import seaborn as sns    #----------------> for inbuilt titanic data set
from scipy.stats import chi2_contingency #---------->for chi2 functions
import pandas as pd   
```

```python
df = sns.load_dataset('titanic')
df
```

```python
contingency_table = pd.crosstab(df['sex'],df['survived']) #------------> or observed frequency
contingency_table
```

```python
chi2, p_value , dof , expected = chi2_contingency(contingency_table) #----> this fn return 4 values a tuple

print("Chi square value = ",chi2)
print("p value = ", p_value)
print("degree of freedom = ", dof)
print("Expected frequency = ",expected)
```

```python
alpha = 0.05     #------> significant value given
```

```python
if p_value < alpha:
    print("we reject the null hypothesis and there is a significant relationship between gender and survivl")
else:
    print("we accept the null hypothesis , there is no connenction")
```
