```python
import numpy as np
from scipy import stats
```

```python
sample = [
    61, 58, 62, 59, 65, 60, 56, 63, 61, 64,
    55, 62, 59, 60, 63, 58, 61, 62, 60, 57,
    59, 64, 61, 63, 58
]

pop_mean = 60
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)
alpha = 0.05   #----------------------> Significant value
```

```python
# calculating t__stat
t_stats = (sample_mean - pop_mean)/(sample_std/np.sqrt(n))
t_stats
```

```python
# calculating p_value
p_value = 2 * (1 - stats.t.cdf(abs(t_stats), df = n-1))
p_value
```

```python
if p_value < alpha:
    print("Reject the null hypothesis")
else: 
    print("Accept the null hypothesis")
```
