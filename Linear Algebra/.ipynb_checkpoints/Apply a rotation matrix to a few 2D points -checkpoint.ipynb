```python
import numpy as np
import matplotlib.pyplot as plt

#sample 2D points
points = np.array([
    [3, 0],
    [0, 3],
    [-2, 2],
    [2, -1]
])


# Rotation matrix for 45 degrees (counter-clockwise)
theta = np.radians(45)                        #-----> Convert 45 degrees to radians
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])


# Apply rotation to all points
rotated = points @ R.T                  #---------> (R.T because points are row vectors)


# Plot setup
fig, ax = plt.subplots(figsize=(8, 8)) #-----> Create figure and axes


# Plot original and rotated points
ax.scatter(points[:, 0], points[:, 1], color='blue', label='Before')
ax.scatter(rotated[:, 0], rotated[:, 1], color='red', label='After rotation (45°)')


# Draw lines from origin to each point
for p, r in zip(points, rotated):
    ax.plot([0, p[0]], [0, p[1]], color='blue', alpha=0.4)       #----> Original lines
    ax.plot([0, r[0]], [0, r[1]], color='red', alpha=0.4)        #-----> Rotated lines

    
# Add x and y axis lines
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)


# Set plot limits and style
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')         #----> Keep x and y scale equal


# Add labels and title
ax.legend()
ax.set_title('Points Before and After 45° Rotation')
plt.grid(True, alpha=0.3)        #---->Add light grid
plt.show()                       #-----> Display the plot
```
