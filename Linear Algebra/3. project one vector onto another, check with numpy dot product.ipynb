```python
import numpy as np
import matplotlib.pyplot as plt

# The vector we are projecting ONTO (think of this as a straight line/direction)
line_vector = np.array([3, 1])

# The vector we want to project
original_vector = np.array([2, 3])

# Step 1:  (dot product)
dot_product = np.dot(original_vector, line_vector)

# Step 2: how long is the line vector, squared (dot product with itself)
length_of_line_squared = np.dot(line_vector, line_vector)

# Step 3: this tells us how many times we need to scale the line vector
# to reach the projection point
scale_amount = dot_product / length_of_line_squared

# Step 4: the actual projected vector
projected_vector = scale_amount * line_vector

print("Dot product (original . line):", dot_product)
print("Length of line vector squared:", length_of_line_squared)
print("Scale amount:", scale_amount)
print("Projected vector:", projected_vector)

# Check: leftover part of original_vector should be perpendicular to line_vector
leftover_vector = original_vector - projected_vector
check_value = np.dot(leftover_vector, line_vector)
print("Check (should be close to 0):", check_value)

# ----> Draw the picture ----
figure, axes = plt.subplots(figsize=(6, 6))

def draw_arrow(vector, color, name):
    axes.quiver(0, 0, vector[0], vector[1],
                angles='xy', scale_units='xy', scale=1,
                color=color, label=name, width=0.008)

draw_arrow(line_vector, 'blue', 'line_vector')
draw_arrow(original_vector, 'red', 'original_vector')
draw_arrow(projected_vector, 'green', 'projected_vector')

# dashed line connecting original_vector to its projection
axes.plot([original_vector[0], projected_vector[0]],
          [original_vector[1], projected_vector[1]],
          'k--', alpha=0.5)

axes.axhline(0, color='gray', linewidth=0.5)
axes.axvline(0, color='gray', linewidth=0.5)
axes.set_xlim(-1, 4)
axes.set_ylim(-1, 4)
axes.set_aspect('equal')
axes.grid(True, alpha=0.3)
axes.legend()
axes.set_title('Projecting one vector onto another')
plt.show()
```
