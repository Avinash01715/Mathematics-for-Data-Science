```python
import sympy as sp

x = sp.Symbol('x')

# 5 composite functions -- each one is "a function of a function"
# chain rule: d/dx[f(g(x))] = f'(g(x)) * g'(x)

composite_functions = [
    (x**2 + 1)**3,          # 1
    sp.sin(x**2),            # 2
    sp.exp(2*x + 1),         # 3
    sp.sqrt(x**2 + 3),        # 4
    sp.cos(sp.sin(x))         # 5
]

for fun in composite_functions:
    derivative = sp.diff(fun, x)
    print(f"f(x) = {fun}---------> f'(x) = {derivative}")
    print("-" * 100)
```
