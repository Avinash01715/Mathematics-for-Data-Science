```python
import sympy as sp

#---> define x as Symbol/variable
x = sp.Symbol('x')


func_list = [
     x**2,            # 1. power rule
     x**3 + 2*x,      # 2. power rule + sum rule
     sp.sin(x),       # 3. trig derivative
     sp.exp(x),       # 4. exponential derivative
     x**2 * sp.sin(x) # 5. product rule
]


# go through all the functions of list
# and find derivative of each
for func in func_list:
    derivative = sp.diff(func, x)                         #--------> differentiate with respect to x
    print(f"f(x) = {func} --------->  f'(x) = {derivative} ")
    print("-" *50)
```
